## Time Series Analysis

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# What are the average, minimum and maximum values?
```

```python
time_series = pd.read_csv('https://tubcloud.tu-berlin.de/s/nwCrNLrtL6LAN3W/download/time-series-lecture-2.csv', index_col=0)
time_series
```

```python
# Find the timestamps where data on prices is missing.
```

```python
missing_timestamps = time_series[time_series['prices [€/MWh]'].isna()].index
print(missing_timestamps)
```

```python
# Fill up the missing data with the prices observed one week ahead in the DataFrame. I.e. for 
# missing data on 1 January 12:00, look for the values on 8 January 12:00
```

```python
time_series.index = pd.to_datetime(time_series.index)
shifted_prices = time_series['prices [€/MWh]'].shift(periods=-7, freq='D').fillna(method='ffill')
time_series['prices [€/MWh]'] = time_series['prices [€/MWh]'].fillna(shifted_prices)
time_series
```

```python
# Plot the time series for the full year.
```

```python
time_series.plot(y="load [GW]", figsize=(20, 5))
plt.xlabel("Timestamp",  size = 10)
plt.ylabel("Load [GW]", size = 10)
plt.title("Time Series for Full Year", size = 15)
```

```python
# Plot the time series for the month May
```

```python
time_series.loc['2015-05-01':'2015-05-31'].plot(y="load [GW]", figsize=(20, 5))
plt.xlabel("Timestamp",  size = 10)
plt.ylabel("Load [GW]", size = 10)
plt.title("Time Series for Month of May", size = 15)
```

```python
# Resample the time series to daily, weeky, and monthly frequencies and create a plot that contains
# the variously resampled time series in one figure.
```

```python
time_series.index = pd.to_datetime(time_series.index) #set index to DateTimeIndex

time_series_daily = time_series.resample('D').mean()
time_series_weekly = time_series.resample('W').mean()
time_series_monthly = time_series.resample('M').mean()

fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=True) # Create a figure with three subplots

time_series_daily.plot(ax=axs[0], title='Daily')
time_series_weekly.plot(ax=axs[1], title='Weekly')
time_series_monthly.plot(ax=axs[2], title='Monthly')

axs[2].set_xlabel('Date')

plt.show()
```

```python
# Sort the values in descending order and plot the duration curve. Hint: Run (g) .reset_index(drop=True)
# to drop the index after sorting.
```

```python
duration_curve = time_series.sort_values('load [GW]', ascending=False)
duration_curve.reset_index(drop=True)
duration_curve['load [GW]'].plot(use_index=False, xlabel='hours [hrs]', ylabel='load [GW]', figsize=(20, 5))
```
